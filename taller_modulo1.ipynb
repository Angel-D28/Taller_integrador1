### 1.1 — Listas y operaciones básicas

Tienes la siguiente lista con la esperanza de vida de 10 países (en años).

````python
# Lista dada — NO modificar
esperanzas_de_vida = [72.3, 81.5, 60.1, 78.9, 55.4, 83.2, 67.8, 74.0, 49.3, 79.6]
````

**Ejercicio 1.1** — Usando Python puro (sin Pandas ni funciones built-in como `sum()` o `max()`):
- Calcula el **promedio** de la lista usando un bucle `for`
- Encuentra el **valor máximo** usando un bucle `for`
- Encuentra el **valor mínimo** usando un bucle `for`

````python
# Tu código aquí

# Promedio
sumatoria = 0
for vida in esperanzas_de_vida:
    sumatoria += vida
promedio = sumatoria/len(esperanzas_de_vida)
print(f"Promedio: {promedio}")

# Máximo
maximo = esperanzas_de_vida[0]; 
for vida in esperanzas_de_vida:
    if vida > maximo:
        maximo =vida
print(f"Maximo: {maximo}")

# Mínimo
minimo = esperanzas_de_vida[0]; 
for vida in esperanzas_de_vida:
    if vida < minimo:
        minimo =vida
print(f"Mínimo: {minimo}")
````

### 1.3 — Funciones

Las funciones nos permiten reutilizar lógica.

**Ejercicio 1.3** — Crea una función llamada `clasificar_esperanza_vida` que reciba un número (años) y retorne:
- `"Alta"` si es >= 75
- `"Media"` si es entre 60 y 74.9
- `"Baja"` si es < 60

Prueba la función con al menos 5 valores distintos e imprime los resultados.

````python
# Tu código aquí
def clasificar_esperanza_vida(años):
    if años >= 75:
        return "Alta"
    elif años >= 60 and años <= 74.9:
        return "Media"
    else:
        return "baja"


valores = [40 , 80, 90 , 72.7 , 60]

for valor in valores:
    clasificacion = clasificar_esperanza_vida(valor)
    print(f"{valor} -> esperanza de vida {clasificacion}")



````

### 2.2 — Limpieza de datos

🔧 **Ejercicio 2.2** — El dataset tiene algunos valores nulos (¡datos reales siempre tienen este problema!).  
1. Rellena los valores nulos de columnas **numéricas** con la **media** de cada columna
2. Verifica que ya no queden nulos
3. Explica en una celda de texto: ¿Por qué usamos la media y no simplemente eliminamos esas filas?

````python
# Tu código aquí

#seleccione solo las columnas con valores numericos; ignore pais , region y nivel de ingresos
columnas_numericas = sm.select_dtypes(include = 'number').columns

for col in columnas_numericas:
    media_col = sm[col].mean()
    sm[col] = sm[col].fillna(media_col)

print(f"Nulos: {sm.isnull().sum().sum()}")

sm.isnull().sum()
````

**Responde aquí** (doble clic para editar):

*¿Por qué usamos la media para rellenar nulos en vez de eliminar las filas?*

> Tu respuesta: Si eliminamos la fila tan solo por tener  un vaalor nulo estariamos perdiendo informacion valiosa a la hora de un analisis  de datos, Esto podria generarr sesgos en el analisis y generarnos una conclusion erronea. Por ejemplo, Angola solo tiene NaN en esperanza de vida, pero el resto de sus datos como la obesidad , medicos , vacunacion, etc. sigen siendo demasiado utiles. Rellenar con la media es una estrategia que nos permite llenar esos espacios sin distorsionar el resto de la información, lo cual nuestro dataset se mantendra estable.
